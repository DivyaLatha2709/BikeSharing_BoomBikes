## Dummy variables


```python
season1 = pd.get_dummies(bike_data['season'])
season1.head()
```

As you can see if season 2,3,4 are all zeros we can say it as season is 1. So we can drop the column 1

```python
weather1 = pd.get_dummies(bike_data['weathersit'])
weather1.head()
```

```python
weekday1 = pd.get_dummies(bike_data['weekday'])
weekday1.head()
```

```python
bike_data.mnth.value_counts()
```

```python
mnth1 = pd.get_dummies(bike_data['mnth'])
mnth1.head()
```

```python
# Let's drop the first column from season1 df using 'drop_first = True'

season1 = pd.get_dummies(bike_data['season'], drop_first = True)
season1.head()
```

```python
# Let's drop the first column from weekday1 df using 'drop_first = True'

weekday1 = pd.get_dummies(bike_data['weekday'], drop_first = True)
weekday1.head()
```

```python
# Let's drop the first column from mnth1 df using 'drop_first = True'

mnth1 = pd.get_dummies(bike_data['mnth'], drop_first = True)
mnth1.head()
```

```python
# Add the results to the original dataframe

bike_data = pd.concat([bike_data, season1], axis = 1)
bike_data = bike_data.rename(columns={2:"Summer",3:"fall",4:"Winter"})
bike_data = pd.concat([bike_data, weather1], axis = 1)
bike_data = bike_data.rename(columns = {1:"weathersit_1",2:"weathersit_2",3:"weathersit_3"})

bike_data.head()
```

```python
bike_data = pd.concat([bike_data, weekday1], axis = 1)
bike_data = bike_data.rename(columns={1:"Tuesday",2:"Wednesday",3:"Thursday",4:"Friday",5:"Saturday",6:"Sunday"})
bike_data.head()
```

```python
bike_data = pd.concat([bike_data, mnth1], axis = 1)
bike_data = bike_data.rename(columns={2:"Feb",3:"Mar",4:"April",5:"May",6:"June",7:"July", 8:"Aug", 9:"Sep",10:"Oct",11:"Nov",12:"Dec"})
bike_data.head()
```

```python
# Drop 'season' as we have created the dummies for it

bike_data.drop(['season'], axis = 1, inplace = True)
bike_data.drop(['weathersit'], axis=1, inplace = True)
bike_data.drop(['weekday'], axis=1, inplace=True)
bike_data.drop(['mnth'], axis=1, inplace=True)
```

```python
bike_data.head()
```

## Splitting the cells into training and test data sets

```python
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
df_train , df_test = train_test_split(bike_data, train_size=0.7, test_size =0.3, random_state =100)
```

### Rescaling the Features

```python
bike_data.cnt.max()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
# Applying scaler() to all the columns except the 'categorical' and 'dummy' variables
num_vars = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

So, we pick `temp` as the first variable and we'll try to fit a regression line to that.

## Dividing X and Y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

## Building a linear model

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train[['temp']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Checking the parameters obtained

lr.params
```

```python
# Let's visualise the data with a scatter plot and the fitted regression line
plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], 0.127 + 0.462*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
# Printing summary of the linear regression model obtained
print(lr.summary())
```

## Building our model using RFE


```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building model using statsmodel

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

#### atemp is the insignificant variable since it has higher P value. So it can be dropped

```python
X_train_rfe.columns
```

```python
X_train_new = X_train_rfe.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### We generally need VIF value less than 5. So dropping the columns one by one and observing the change

```python
X_train_new1 = X_train_new.drop(["hum"], axis = 1)
```

```python
X_train_lm1 = sm.add_constant(X_train_new1)
lm = sm.OLS(y_train,X_train_lm1).fit()
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new1
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Residual Analysis of the train data

```python
y_train_cnt = lm.predict(X_train_lm1)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18) 
```

## Making Predictions

#### Applying the scaling on the test sets

```python
num_vars = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']

df_test[num_vars] = scaler.transform(df_test[num_vars])
```

## Dividing into x_test and y_test

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_new1.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_pred = lm.predict(X_test_new)
```

### Model Evaluation

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)    
```

We can see that the equation of our best fitted line is:

$ cnt = 0.2335  \times  yr - 0.0853  \times  holiday  -0.2526 \times weathersit_3 - 0.1426 \times windspeed + 0.0895 \times summer + 0.5471 \times temp + 0.1330 \times winter + 0.0390 \times Aug + 0.0997 \times Sep $

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

Overall we have a decent model, but we also acknowledge that we could do better. 

The difference in R squared is 0.032 which is comparitively less
